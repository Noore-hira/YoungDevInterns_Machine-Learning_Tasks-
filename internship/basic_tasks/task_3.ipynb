## Visualize Data with a Scatter Plot

```python
#importing neccessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading Iris dataset from seaborn library
df=sns.load_dataset('iris')
df.head()
```

```python
sns.countplot(x='species', data=df, palette='rocket')
plt.show()
```

```python
#Seeing relation between Sepal Length and Sepal Width
sns.scatterplot(x='sepal_length', y='sepal_width',hue='species', data=df )
```

```python
# seeing relation between Petal Length and Petal Width
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df )
```

### Let’s plot all the column’s relationships using a pairplot. It can be used for multivariate analysis.

```python
sns.pairplot(df, hue='species', height=2)
```

## Analyzing Features Corelation using Heatmaps

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(method='pearson'), annot = True) 
```
