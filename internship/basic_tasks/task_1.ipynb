## Implement a Simple Linear Regression

```python
#importing neccessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
#loading dataset of tips from seaborn library
df=sns.load_dataset('tips')
print('First 5 Rows:\n',df.head())
print('Shape of Dataset: ',df.shape)
print('Statistical Operations on Dataset:\n',df.describe())
```

```python
#separating X and y
X = df[['total_bill']]
y = df['tip']
```

```python
#splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#fitting Linear Regression model
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#predicting y for X_test
y_pred = model.predict(X_test)

#Evaluating model 
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred)))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
#plotting the Regression line
plt.scatter(X_test,y_test)
plt.plot(X_test, y_pred, color='red')
plt.show()
```

```python
#Make predictions using model
print(model.predict([[1000]]))
print(model.predict([[100]]))
print(model.predict([[2678]]))
print(model.predict([[3700]]))
print(model.predict([[9000]]))
```
